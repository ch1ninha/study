### Julho 04 | Minimum X (xor) A
https://practice.geeksforgeeks.org/problems/x-xor-a-is-minimum-and-set-bits-in-x-b/1

```python
#User function Template for python3

class Solution:
    def countBits(self,num):
        return num.count("1")
    
    def minVal(self, a, b):
        b_bin = bin(b)
        a_bin = bin(a)
        
        x = 1
        dic_xor = {}
        
        if self.countBits(a_bin) == self.countBits(b_bin):
            return a
        while x != 0:
            x_bin = bin(x)
            # print("x:",x)
            # print("xor (x ^ a):",x ^ a)
            if self.countBits(x_bin) == self.countBits(b_bin):
                # print("ativou qtde")
                # print(f"(x ^ a): {(x ^ a)} xor_pequeno: {xor_pequeno})")
                xor_pequeno = (x ^ a)
                dic_xor[xor_pequeno] = x
            x += 1
            
            if x > (a|b) and len(dic_xor) > 0:
                break
            # print("--------")
        min_dic = min(dic_xor.keys())
        return dic_xor[min_dic]
```

## Count the Substrings | Problema 2

https://practice.geeksforgeeks.org/contest/the-easiest-ever-coding-challenge-3/problems/

```python
class Solution:
    def countSubstring(self, S): 
        lista_substrings = [] # adicionar as substrings que existirem
        for j in range(len(S),0,-1):
            for i in range(len(S)):
                cond1 = Solution.qtdeMai(S[i:j])
                cond2 = Solution.qtdeMin(S[i:j])
                if cond1 == cond2 and S[i:j] != "":
                    lista_substrings.append(S[i:j])
                    continue
        
        return len(lista_substrings)
        
    @staticmethod
    def qtdeMai(s):
        qtde_maiusculo = 0
        for char in s:
            if char.isupper():
                qtde_maiusculo += 1
        return qtde_maiusculo
    
    @staticmethod    
    def qtdeMin(s):
        qtde_minusculo = 0
        for char in s:
            if char.islower():
                qtde_minusculo += 1
        return qtde_minusculo
```

## Problem Setting at GFG | Problema 5 

https://practice.geeksforgeeks.org/contest/the-easiest-ever-coding-challenge-3/problems/

```python
from typing import List


class Solution:
    def maximum_problems(self, n : int, problems : List[int]) -> int:
        lista_problema = []
        tamanho_maior_problema = 0
        
        if len(problems) <= 1:
            return 1

        for i in range(len(problems)):
            # checa se não chegou no ultimo num
            if i == n-1:
                if (problems[i-1] * 2) >= problems[i]:
                    lista_problema.append(problems[i])
                
                tam_problema_atual = len(lista_problema)
                if tam_problema_atual > tamanho_maior_problema:
                    return tam_problema_atual
                break
            #     break
            a0 = problems[i]
            a1 = problems[i+1]
            if (a0 * 2) >= a1:
                # print(f"({a0} * 2) > {a1}")
                # OK, continua
                lista_problema.append(a0)
                continue
            else:
                lista_problema.append(a0)
                tam_problema_atual = len(lista_problema)
                if tam_problema_atual > tamanho_maior_problema:
                    tamanho_maior_problema = tam_problema_atual
                lista_problema = []
        return tamanho_maior_problema
```

## Print the pattern | Set-1

https://practice.geeksforgeeks.org/problems/print-the-pattern-set-1/1

```python
def printPat(n):
    repeticoes = n
    while repeticoes:
        for i in range(n,0,-1):
            for j in range(repeticoes):
                print(i, end=" ")
        print("$", end="")
        repeticoes -= 1
```

## Second Largest

https://practice.geeksforgeeks.org/problems/second-largest3735/1

```python
class Solution:
    def print2largest(self,arr, n):
        primeiro_valor = 0
        segundo_valor = 0
        for i in range(n):
            if arr[i] > primeiro_valor:
                segundo_valor = primeiro_valor
                primeiro_valor = arr[i]
            elif arr[i] > segundo_valor and arr[i] < primeiro_valor:
                segundo_valor = arr[i]
        if segundo_valor == 0:
            return -1
        return segundo_valor
```

## Find the median

https://practice.geeksforgeeks.org/problems/find-the-median0527/1

```python
class Solution:
    def find_median(self, v):
        def mergeSort(arr):
            if len(arr) > 1:
                mid = len(arr) // 2
                l = arr[:mid]
                r = arr[mid:]
                
                mergeSort(l)
                mergeSort(r)
                i = j = k = 0
                while i < len(l) and j < len(r):
                    if l[i] < r[j]:
                        arr[k] = l[i]
                        i += 1
                    else:
                        arr[k] = r[j]
                        j += 1
                    k += 1
                while i < len(l):
                    arr[k] = l[i]
                    i += 1
                    k += 1
                    
                while j < len(r):
                    arr[k] = r[j]
                    j += 1
                    k += 1
            return arr
        arr = mergeSort(v)
        
        meio_arr = len(arr) // 2
        if len(arr) % 2 == 0:
            return (arr[meio_arr] + (arr[meio_arr-1])) // 2
        else:
            return arr[meio_arr]
```
